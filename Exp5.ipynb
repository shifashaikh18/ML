```python
import numpy as np

def hebbian_update(inputs, output, weights, bias, learning_rate=1):
    """
    Updates weights and bias using the simple Hebbian Learning rule.
    """
    # Update each weight
    for i in range(len(weights)):
        weights[i] += learning_rate * inputs[i] * output

    # Update bias
    bias += learning_rate * output

    return weights, bias


# Training data for AND gate (bipolar)
training_data_and = [
    (np.array([-1, -1]), -1),
    (np.array([-1,  1]), -1),
    (np.array([ 1, -1]), -1),
    (np.array([ 1,  1]),  1)
]

# Initialize weights and bias
num_inputs = 2
weights = np.zeros(num_inputs)
bias = 0
learning_rate = 1

print("AND Gate")
print("Initial weights:", weights)
print("Initial Bias:", bias)

# Training loop
for inputs_vec, target_output in training_data_and:
    weights, bias = hebbian_update(inputs_vec, target_output, weights, bias, learning_rate)
    print(f"Input {inputs_vec}: {weights}, Bias: {bias}")

print("Final AND Weights:", weights)
print("Final AND Bias:", bias)
```
