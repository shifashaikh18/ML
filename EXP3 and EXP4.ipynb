```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Sample dataset (1 feature)
X = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])  # actual values

# Create and train model
model = LogisticRegression()
model.fit(X, y)

# Predicted probabilities
y_pred_prob = model.predict_proba(X)[:, 1]

# Plot actual values
plt.plot(X, y, 'o-', label="Actual Values")

# Plot predicted values
plt.plot(X, y_pred_prob, 'r-', label="Predicted Values")

plt.xlabel("Input Feature (X)")
plt.ylabel("Output")
plt.title("Logistic Regression: Actual vs Predicted")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Main cluster (Class 0)
X1 = np.random.randn(20, 2) * 0.5 + [2, 2]

# Main cluster (Class 1)
X2 = np.random.randn(20, 2) * 0.5 + [6, 6]

# Outside / outlier points
outliers = np.array([
    [0, 7],
    [8, 1],
    [9, 9],
    [1, 1]
])

# Combine data
X = np.vstack((X1, X2, outliers))
y = np.array([0]*20 + [1]*20 + [1]*4)  # label outliers as class 1

# Train SVM
model = SVC(kernel='linear')
model.fit(X, y)

# Plot points
plt.scatter(X1[:,0], X1[:,1], label="Class 0")
plt.scatter(X2[:,0], X2[:,1], label="Class 1")
plt.scatter(outliers[:,0], outliers[:,1], color='red', marker='x', s=100, label="Outside Points")

# Decision boundary
w = model.coef_[0]
b = model.intercept_[0]
x_line = np.linspace(0, 10, 100)
y_line = -(w[0] * x_line + b) / w[1]
plt.plot(x_line, y_line, 'k--', label="Decision Boundary")

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("SVM")
plt.legend()
plt.grid(True)
plt.show()
```
